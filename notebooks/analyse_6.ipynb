```python
def unique_or_error(x):
    uniques = x.dropna().unique()
    if len(uniques) == 1:
        return uniques[0]
    elif len(uniques) == 0:
        return None
    else:
        raise ValueError(f"Multiple unique values found in group: {uniques}")

df = df.groupby(
    ['partenaire', 'contrat', 'activite', 'secteur', 'annee', 'mois'],
    as_index=False
).agg({
    "ventes": "sum",
    "consommation": "sum",
    "temperature": "mean",
    "puissance facturée": "sum",
    "Date d'emménagement": unique_or_error,
    "Date de déménagement": unique_or_error
})
```

```python
import numpy as np
```

```python
raw_source = pd.read_excel(PROJECT_ROOT / "data/raw/Puissances appelées CD.xlsx", sheet_name = "Tarif super pointe")
```

```python
new_columns = pd.MultiIndex.from_arrays(raw_source.iloc[3:5].values)

# Drop the first two rows and set the new columns
raw_source.columns = new_columns

raw_source = raw_source.iloc[5:].reset_index(drop=True)
```

```python
raw_source = raw_source.drop(columns = [(np.nan, 'Business Partner')])
```

```python
raw_source[(np.nan, 'Contrat')] = raw_source[(np.nan, 'Contrat')].astype(int)
```

```python
new_mapping = {
    5729390: 6447659,
    5731079: 6457296,
    5741477: 6434280,
    6192130: 6457387,
    6160312: 6457370,
    217176: 6500531,
    217169: 8192392,
    5331362: 9236601,
    217184:6149701,
}
```

```python
raw_source.replace(new_mapping, inplace=True)
```

```python
len(raw_source)
```

```python
raw_source = raw_source.groupby((np.nan, "Contrat"), as_index = False).sum()
```

```python
import pandas as pd
import numpy as np

# Prepare an empty DataFrame for results
pf_df = pd.DataFrame(index=raw_source.index)

# Loop over each month (first level of MultiIndex)
for month in raw_source.columns.levels[0][:-1]:
    subcols = raw_source[month]

    # Check which tariff columns exist for this month
    cols = subcols.columns

    if set(['HP', 'HPL', 'HC']).issubset(cols):
        # --- Tarif Optionnel ---
        PS1 = subcols['HP']
        PS2 = subcols['HPL']
        PS3 = subcols['HC']
        Pf = PS1 + 0.6*(PS2 - PS1) + 0.4*(PS3 - PS2)

    elif set(['HSP', 'HP', 'HPL', 'HC']).issubset(cols):
        # --- Tarif Super Pointe ---
        PS1 = subcols['HSP']
        PS2 = subcols['HP']
        PS3 = subcols['HPL']
        PS4 = subcols['HC']
        Pf = PS1 + 0.8*(PS2 - PS1) + 0.6*(PS3 - PS2) + 0.4*(PS4 - PS3)

    else:
        # If no known tariff combination found
        Pf = np.nan

    # Add Pf as the only subcolumn for this month
    pf_df[(month, 'Pf')] = Pf

# Sort and clean up columns
pf_df.columns = pd.MultiIndex.from_tuples(pf_df.columns)
pf_df = pf_df.sort_index(axis=1)
```

```python
pf_df
```

```python
raw_source = pd.concat([raw_source[[(np.nan, 'Contrat')]], pf_df], axis=1)
```

```python
raw_source.columns = [
    f"{lvl2}__{lvl1}" if not pd.isna(lvl1) else str(lvl2)
    for lvl1, lvl2 in raw_source.columns
]
```

```python
import re
import pandas as pd

id_cols = ['Contrat']

val_cols = [c for c in wide.columns if re.match(r'^.+__\d{2}\.\d{4}$', str(c))]

long = wide.melt(id_vars=id_cols, value_vars=val_cols,
               var_name='key', value_name='value')

parts = long['key'].str.extract(r'^(?P<metric>.+?)__(?P<Month>\d{2})\.(?P<Year>\d{4})')
long = pd.concat([long.drop(columns='key'), parts], axis=1)
long['Month'] = long['Month'].astype(int)
long['Year'] = long['Year'].astype(int)
long['metric'] = long['metric'].str.strip()

wide = (long.pivot_table(index=id_cols+['Year','Month'],
                         columns='metric', values='value', aggfunc='first')
             .reset_index())

wide = wide.rename(columns={'Pf':'puissance appelée'})
```
