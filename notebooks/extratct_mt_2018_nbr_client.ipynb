```python
def fmt_two_int_digits(val: float) -> str:
    s = f"{val:.4f}"           # e.g. "3.0000" or "-7.1250"
    sign = "-" if s.startswith("-") else ""
    if sign: s = s[1:]
    int_part, frac = s.split(".")
    int_part = int_part.zfill(2)  # ensure at least two digits before the dot
    return f"{sign}{int_part}.{frac}"

cols = ["City", "Activity"]
for val in df.iloc[0, 2:]:
    cols.append(f"NB_CLIENTS__{fmt_two_int_digits(val)}")
```

```python
df.columns = cols
```

```python
import re
import pandas as pd

id_cols = ['City', 'Activity']

val_cols = [c for c in df.columns if re.match(r'^.+__\d{2}\.\d{4}$', str(c))]

long = df.melt(id_vars=id_cols, value_vars=val_cols,
               var_name='key', value_name='value')

parts = long['key'].str.extract(r'^(?P<metric>.+?)__(?P<Month>\d{2})\.(?P<Year>\d{4})')
long = pd.concat([long.drop(columns='key'), parts], axis=1)
long['Month'] = long['Month'].astype(int)
long['Year'] = long['Year'].astype(int)
long['metric'] = long['metric'].str.strip()

wide = (long.pivot_table(index=id_cols+['Year','Month'],
                         columns='metric', values='value', aggfunc='first')
             .reset_index())

wide = wide.rename(columns={'NB_CLIENTS':'Nbr Clients'})
```

```python
keys = ['City', 'Activity', 'Year', 'Month']          # common columns
target_col = 'Nbr Clients'                            # column to fill in final_df
source_col = 'Nbr Clients'                            # column in wide that has the values

# build source series (in case wide has duplicates, pick one rule: mean)
src = (wide
       .dropna(subset=[source_col])
       .groupby(keys, as_index=True)[source_col]
       .mean())                                       # or .first(), .max(), etc.
print(len(src))
# merge and fill
final_df = final_df.merge(src.rename(f'{target_col}_src'),
                          on=keys, how='left')

final_df[target_col] = final_df[target_col].fillna(final_df[f'{target_col}_src'])
final_df = final_df.drop(columns=[f'{target_col}_src'])
```
