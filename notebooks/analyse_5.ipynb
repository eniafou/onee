```python
import pandas as pd
from kmodes.kprototypes import KPrototypes
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np

year = 2023
# Example: assume your df is called df
# Keep only the two columns we care about
data = df[df["annee"] == year][['contrat', 'activite', 'mois', 'puissance facturée']].copy().dropna()

# Convert categorical column to string if needed
data['activite'] = data['activite'].astype(str)
```

```python
da = data[["contrat", "activite"]].drop_duplicates()
```

```python
dd = data.pivot(index='contrat', columns='mois', values='puissance facturée')
```

```python
data = dd.merge(da, on = "contrat")
```

```python
data["contrat"] = data["activite"]
data.drop(columns = ["activite"], inplace = True)
data.rename(columns = {"contrat":"activite"}, inplace = True)
```

```python

# Prepare data for K-Prototypes
# K-Prototypes expects numpy array and needs to know which columns are categorical
X = data.to_numpy()
print(X.shape)
# Index of categorical columns
categorical_cols = [0]  # 'activite' is the first column

# Initialize K-Prototypes
n_clusters = 6
kproto = KPrototypes(n_clusters=n_clusters, init='Huang', random_state=42)

# Fit the model
clusters = kproto.fit_predict(X, categorical=categorical_cols)

# Add cluster labels to dataframe
data['cluster'] = clusters
```

```python
# Evaluate only on the numeric column
numeric = data[data.columns[1:-1]].to_numpy()

score = silhouette_score(numeric, clusters)
print("Silhouette score (on numeric feature only):", score)
```

```python
import matplotlib.pyplot as plt

# Define the range of clusters to test
cluster_range = range(2, 9)
silhouette_scores = []

for n_clusters in cluster_range:
    # Initialize K-Prototypes
    kproto = KPrototypes(n_clusters=n_clusters, init='Huang', random_state=42)
    
    # Fit the model
    clusters = kproto.fit_predict(X, categorical=categorical_cols)
    
    # Add cluster labels to your dataframe
    data['cluster'] = clusters
    
    # Evaluate only on numeric features
    numeric = data[['puissance facturée']].to_numpy()
    score = silhouette_score(numeric, clusters)
    
    silhouette_scores.append(score)
    print(f"Clusters: {n_clusters}, Silhouette Score: {score:.4f}")

# Plot the silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters (K-Prototypes)')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
from onee.utils import safe_parse_date
```

```python
def get_move_in_year(df, not_started_yet_th = 20):
    """
    Extract the first non-null 'Date d'emménagement' value from a DataFrame
    and return its year as an integer, or None if not available.
    """
    if 'Date d\'emménagement' not in df.columns:
        return None

    move_in_val = df['Date d\'emménagement'].dropna()
    if move_in_val.empty:
        return None

    from datetime import datetime
    move_in_date = safe_parse_date(move_in_val.iloc[0])
    if move_in_date is None:
        return None
    move_in_year = move_in_date.year
    
    max_year = df["annee"].max()
    annual_consumption = df[df["annee"] == move_in_year]["consommation"].sum()
    while annual_consumption < not_started_yet_th:
        move_in_year+=1
        if move_in_year >= max_year:
            break
        annual_consumption = df[df["annee"] == move_in_year]["consommation"].sum()

    return move_in_year
```
